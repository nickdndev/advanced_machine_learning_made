## 1.1
_Произошло убийство. На месте убийства найдена кровь, которая явно (будем считать, что с вероятностью 1) принадлежит настоящему убийце. Кровь принадлежит редкой группе, которая присутствует всего лишь у 1% населения. И так получилось, что у подсудимого как раз эта редкая группа крови!\
Первым выходит прокурор и говорит: “Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен -- всего 1%; значит, с вероятностью 99% он виновен, предлагаю осудить”. В чём не прав прокурор? Укажите, какие вероятности он оценил, а какие должен был оценить._


### Решение:

$A$- виновный;
$\overline{A}$ - невиновный;
$B$ - редкая группа крови;
$\overline{B}$ - нередкая группа крови.


- $P(B) = 0,01$ -  вероятность редкой группы.
- $P(B|A) = 1$ - вероятность наличия у человека редкой группы крови при условии, что он убийца.

**Вывод:**

**Прокурор неправильно оценил вероятность виновности подсудимого при условии редкой группы крови $P(A|B) = 1 -P(\overline{A}|B) = 1-0.01 = 0.99$, он принял  вероятность невиновности при условии редкой группы крови $P(\overline{A}|B)$ за 0.01, хотя 0.01 это вероятность редкой группы крови $P(B)$.**


## 1.2

_Затем выходит адвокат, объясняет ошибки прокурора и берёт слово сам: “В городе живёт миллион человек. Это значит, что примерно у 10000 из них такая группа крови. Значит, всё, что говорит нам эта кровь -- это то, что подсудимый совершил убийство с вероятностью 0.01%. Получается, что это свидетельство не просто не является строгим доказательством, а и вовсе даёт нам ничтожно малую вероятность того, что мой подзащитный виновен, так что предлагаю исключить его из материалов дела”. В чём не прав адвокат? Какие вероятности он оценил, а какие должен был оценить?_

### Решение:

**Главная цель адвоката защита подсудимого, но адвокат акцентировал внимание на небольшую вероятность виновности при условии редкой группы крови. Но необходимо обратить внимание на разницу априорной (0.0001%) и апостериорной вероятноти (0.01%) и она получается в 100 раз больше.\
И видно ,что исключать улику о группе крови нельзя т.к. увеличивает вероятность виновности подсудимого в сотни раз.**

$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} =\frac{1 \cdot \frac{1}{1000000}}{\frac{10000}{1000000}} = \frac{1 \cdot 0.000001}{0,01} = 0.0001$ или 0.01%

## 1.3

_Вернёмся к тесту на страшную болезнь из первой лекции. Напомню, что по предположениям задачи болеет 1% населения, а тест имеет вероятность ошибки 5% в обе стороны. Мы видели, что в результате человек с положительным тестом имеет апостериорную вероятность оказаться больным всего лишь около 16%.\
Врачи решили, что терпеть это никак нельзя; они готовы провести срочное исследование и улучшить качество теста. Но сил и денег хватит только для того, чтобы сократить один тип ошибок, т.е. уменьшить или только число false positives (когда тест выдаёт положительный результат у здорового человека), или только число false negatives (когда тест выдаёт отрицательный результат у больного).\
Помогите медикам: постройте зависимости апостериорной вероятности болезни после положительного и отрицательного теста от обоих видов ошибок (например, в виде графиков), сделайте выводы и дайте рекомендации врачам -- на чём лучше сосредоточиться?_

### Решение:

$A$ = 0 - здоров;
$A$ = 1- болен;
$B$ = 0 - отрицательный тест;
$B$ = 1 - положительный тест.

$P(A = 1|B = 1) = \frac{P(B = 1|A = 1) \cdot P(A = 1)}{P(B = 1)}$
 где: $P(B = 1)= P(B = 1|A = 1) \cdot P(A = 1) + P(B = 1|A = 0) \cdot P(A = 0)$
   
$P(A = 1|B = 0) = \frac{P(B = 0|A = 1) \cdot P(A = 1)}{P(B = 0)} $
 где: $P(B = 0)= P(B = 0|A = 1) \cdot P(A = 1) + P(B = 0|A = 0) \cdot P(A = 0)$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def test_positive(false_negative,false_positive,sick_p):
    return sick_p * (1 - false_negative) / ((1 - false_negative) * sick_p + false_positive * (1 - sick_p))

def test_negative(false_negative,false_positive,sick_p):
    return sick_p * false_negative / (false_negative * sick_p + (1 - false_positive) * (1 - sick_p))
```

```python
PRIORI_PROBABILITY = 0.01
RANGE = np.linspace(0, 0.05, 1000)
FING_SIZE = (6,4)

probabality_posterior_positive = test_positive(false_negative = 0.05,false_positive = RANGE,sick_p = PRIORI_PROBABILITY)

plt.figure(figsize=FING_SIZE)
plt.title('Зависимость  вероятности болезни (false positive)', fontsize=16)
plt.plot(RANGE, probabality_posterior_positive, label='Положительного тест')
plt.legend(fontsize=10)
plt.xlabel('False positive', fontsize=16)
plt.ylabel('Вероятность болезни', fontsize=16);

probabality_posterior_negative = test_negative(false_negative = 0.05,false_positive = RANGE,sick_p = PRIORI_PROBABILITY)

plt.figure(figsize=FING_SIZE)
plt.title('Зависимость вероятности болезни (false positive)', fontsize=16)
plt.plot(RANGE, probabality_posterior_negative, label='Отрицательного теста')
plt.legend(fontsize=10)
plt.xlabel('False positive', fontsize=16)
plt.ylabel('Вероятность болезни', fontsize=16);


probabality_posterior_positive = test_positive(false_negative = RANGE, false_positive = 0.05,sick_p = PRIORI_PROBABILITY)

plt.figure(figsize=FING_SIZE)
plt.title('Зависимость вероятности болезни (false negative)',fontsize=16)
plt.plot(RANGE, probabality_posterior_positive, label='Положительного теста')
plt.legend(fontsize=10)
plt.xlabel('False negative', fontsize=16)
plt.ylabel('Вероятность болезни', fontsize=16);

probabality_posterior_negative = test_negative(false_negative = RANGE, false_positive = 0.05, sick_p = PRIORI_PROBABILITY)

plt.figure(figsize=FING_SIZE)
plt.title('Зависимость вероятности болезни (false negative)',fontsize=16)
plt.plot(RANGE, probabality_posterior_negative, label='Отрицательного теста')
plt.legend(fontsize=10)
plt.xlabel('False negative', fontsize=16)
plt.ylabel('Вероятность болезни', fontsize=16);

```

**Вывод:**

**В период пандемии необходимо в короткие сроки улучшить качество тестов на ковид. Проанализировав моделирования изменения FP и FN видно, что изменение FP имеет экспоненциальную зависимость, что позволяет за короткий период времени при "незначительном" уменьшении FP получить значительное улучшение "качества" тестов.  Также при уменьшении FP позволяет снимать нагрузку на врачей и сконцентрировать все силы на больных.**

Линейная модель будет построена аналитически. В качестве источника теоритических выкладок использовал книгу [Christopher M. Bishop_Pattern Recognition and Machine Learning_](https://www.amazon.com/Pattern-Recognition-Learning-Information-Statistics/dp/0387310738)

```python
class LinearRegression:
    def __init__(self,mean_0,covariance_0):
        self.mean = mean_0
        self.covariance = covariance_0

    def fit(self, X, y):
        X = np.hstack((np.ones((len(X), 1)), X))
        y = y.reshape(-1, 1)

        self.variance_noise = np.var(X.dot(np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)) - y)
        
        beta = 1 / self.variance_noise
        
        refresh_covariance = np.linalg.inv(np.linalg.inv(self.covariance) + beta * X.T.dot(X))

        self.mean = refresh_covariance.dot(np.linalg.inv(self.covariance).dot(self.mean.reshape(-1, 1))+ beta* X.T.dot(y)).flatten()

        self.covariance = refresh_covariance

    def sampling(self, X, n_samples=None):
        X = np.hstack((np.ones((len(X), 1)), X))
        if n_samples is None:
            return X.dot(sps.multivariate_normal.rvs(self.mean, self.covariance))
        else:
            return X.dot(sps.multivariate_normal.rvs(self.mean, self.covariance, size=n_samples).T)
        
    def predict(self, X):
        return np.hstack((np.ones((len(X), 1)), X)).dot(self.mean)
```

```python
lr_total_cases = LinearRegression(np.zeros(2),np.eye(2))
lr_total_cases.fit(train_data_rus_df[["day"]].values, np.log(train_data_rus_df["total_cases"].values))

lr_new_cases = LinearRegression(np.zeros(2),np.eye(2))
lr_new_cases.fit(train_data_rus_df[["day"]].values, np.log(train_data_rus_df["new_cases"].values))
```

```python
locations = [('Germany','2020-03-01'),('Kuwait','2020-03-03'),('Kenya','2020-03-13'),('Spain','2020-02-24'),
             ('Italy','2020-02-24'),('Greece','2020-02-26'),('Chile','2020-03-20'),
             ('Austria','2020-02-25'),('Turkey','2020-03-11'),('Monaco','2020-03-10'),('Estonia','2020-03-03'),
             ('Egypt','2020-02-03'),('Russia','2020-03-03'),('Ukraine','2020-03-03'),('India','2020-02-17'),
             ('United Kingdom','2020-02-20')
            ]
```

```python
models = []
train_size = 50
for location, data_from in locations:
    
    data_temp_df = data_df[
        (data_df["date"] >= data_from) & (data_df["location"] == location)
    ]

    data_temp_df["day"] = (data_temp_df["date"] - pd.to_datetime(date_from)).astype(
        "timedelta64[D]"
    )

    data_temp_df = data_temp_df[:train_size].reset_index(drop=True)


    X_train_sigmoid = np.array([data_temp_df["day"].values, (data_temp_df["day"].values) ** 2]).T


    sigmoid_lin_reg_total_cases = LinearRegression(np.zeros(3),np.eye(3))

    sigmoid_lin_reg_total_cases.fit(
        X_train_sigmoid, np.log(data_temp_df["new_cases"].values)
    )
    
    models.append((location,sigmoid_lin_reg_total_cases.mean))

    plt.figure(figsize=(15, 5))
    plt.plot(data_temp_df["date"], data_temp_df["total_cases"], label="Test data")
    plt.plot(
        data_temp_df["date"],
        np.exp(sigmoid_lin_reg_total_cases.predict(X_train_sigmoid)).cumsum(),
        label="Predictions",
    )

    plt.grid()
    plt.title(f'Total cases for {location}')
    plt.legend()
    plt.show()
```
